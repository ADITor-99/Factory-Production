```python
#Check 1

#Check 1.1

import pandas as pd
 
def check_values_presence(dataframes, column_name, values_to_check):

    """

    Check if all values in the specified column of the dataframes are present in the values_to_check list.

    Parameters:

        dataframes (list): List of pandas DataFrames to check.

        column_name (str): Name of the column to check the values.

        values_to_check (list): List of values to compare against.

    Returns:

        list: List of DataFrames that failed the check.

    """

    failed_dataframes = []
 
    for df in dataframes:

        # Check if any value in the specified column is not present in values_to_check

        if not df[column_name].isin(values_to_check).all():

            failed_dataframes.append(df)
 
    return failed_dataframes
 
# Create a function to generate the output DataFrame

def generate_output_dataframe(dataframes, column_name, values_to_check):

    # Get the names of the dataframes

    table_names = ['df_BOM', 'df_calender', 'df_Factory_Operations', 'df_LabourData', 'df_RM_M1OS',

                   'df_factory', 'df_lines', 'df_RM_Supply', 'df_Production']

    # Initialize an empty dictionary to store the results

    results = {}
 
    for value in values_to_check:

        results[value] = {}
 
        for i, df in enumerate(dataframes):

            results[value][table_names[i]] = value in df[column_name].values
 
        results[value]['Result'] = all(results[value].values())
 
    # Convert the dictionary to a DataFrame

    output_df = pd.DataFrame(results).T.reset_index()

    output_df.columns = [column_name] + table_names + ['Result']
 
    return output_df
 
# Define values and dataframes to check

factory_values_to_check = ['NFC', 'KFC', 'GFC']

factory_dataframes_to_check = [df_BOM, df_calender, df_Factory_Operations, df_LabourData, df_RM_M1OS,

                               df_factory, df_lines, df_RM_Supply, df_Production]
 
# Perform the check

failed_dataframes = check_values_presence(factory_dataframes_to_check, 'FactCode', factory_values_to_check)
 
# Generate the output DataFrame

output_df1_1 = generate_output_dataframe(factory_dataframes_to_check, 'FactCode', factory_values_to_check)
 
# Print the result

if not failed_dataframes and output_df1_1['Result'].all():

    print("Check 1.1 : FactoryCodes are consistent: Pass")

else:

    print("Check 1.1 : FactoryCodes are consistent: Fail")
 
#print(output_df1_1)
 
 
#Check 1.2
 
# Define a function to check the presence of WHCodes

def check_whcodes_presence(base_df, df_dict):

    failing_dataframes = []

    for df_name, wh_codes in df_dict.items():

        if not all(code in base_df for code in wh_codes):

            failing_dataframes.append(df_name)

    return failing_dataframes
 
# Define the base DataFrame

base_wh_codes = set(df_warehouse['WHCode'])
 
# Define WHCodes from other DataFrames

wh_codes_to_check = {

    'df_Secondary_Freight': df_Secondary_Freight['WHCode'].unique(),

    'df_Primary_Freight': df_Primary_Freight['WHCode'].unique(),

    'df_FG_M1OS': df_FG_M1OS['WHCode'].unique()

    #'df_FG_Inventory_Norms': df_FG_Inventory_Norms['WHCode'].unique()

}
 
# Perform the check

failing_dataframes = check_whcodes_presence(base_wh_codes, wh_codes_to_check)
 
# Generate the output DataFrame

def generate_output_dataframe(base_df, check_dict):

    results = {}

    for code in base_df:

        results[code] = {}

        for df_name, wh_codes in check_dict.items():

            results[code][df_name] = code in wh_codes

        results[code]['Result'] = all(results[code].values())

    output_df = pd.DataFrame(results).T

    output_df.index.name = 'WHCode'

    return output_df.reset_index()
 
output_df1_2 = generate_output_dataframe(base_wh_codes, wh_codes_to_check)
 
# Rename the DataFrame columns to match the desired format

output_df1_2.rename(columns={

    'df_Secondary_Freight': 'Secondary_Freight',

    'df_Primary_Freight': 'Primary_Freight',

    'df_FG_M1OS': 'FG_M1OS',

    'df_FG_Inventory_Norms': 'FG_Inventory_Norms',

    'Result': 'Result'

}, inplace=True)
 
# Print Pass or Fail based on the check

if not failing_dataframes:

    print("Check 1.2 : WHCodes are consistent: Pass")

else:

    print("Check 1.2 : WHCodes are consistent: Fail")

    
 
# Display the output DataFrame

#print(output_df1_2)
 
 
#Check1.3
 
# Create a dictionary to store the unique country codes for each DataFrame

unique_country_codes = {

    'df_demand': set(df_demand['CountryCode'].unique()),

#     'df_FG_Restrictions': set(df_FG_Restrictions['CountryCode'].unique()),

    'df_Secondary_Freight': set(df_Secondary_Freight['CountryCode'].unique()),

    'df_country': set(df_country['CountryCode'].unique())

}
 
# Get the base unique country codes from df_demand

base_unique_country_codes = unique_country_codes['df_demand']
 
# Check if all other DataFrames contain only country codes present in df_demand

failed_dataframes = [df_name for df_name, country_codes in unique_country_codes.items() if not country_codes.issubset(base_unique_country_codes)]
 
# Generate the output DataFrame

def generate_output_dataframe(base_codes, check_dict):

    results = {}

    for code in base_codes:

        results[code] = {}

        for df_name, country_codes in check_dict.items():

            results[code][df_name] = code in country_codes

        results[code]['Result'] = all(results[code].values())

    # Check for extra combinations of CountryCode in each DataFrame

    for df_name, country_codes in check_dict.items():

        extra_codes = country_codes - base_codes

        for code in extra_codes:

            results[code] = {}

            for df in check_dict.keys():

                results[code][df] = df == df_name

            results[code]['Result'] = False

    output_df = pd.DataFrame(results).T

    output_df.index.name = 'CountryCode'

    return output_df.reset_index()
 
output_df1_3 = generate_output_dataframe(base_unique_country_codes, unique_country_codes)
 
# Rename the DataFrame columns to match the desired format

output_df1_3.rename(columns={

    'df_demand': 'df_demand',

    'df_Secondary_Freight': 'df_Secondary_Freight',

    'df_country': 'df_country',

    'Result': 'Result'

}, inplace=True)
 
# Print Pass or Fail based on the check

if not failed_dataframes and output_df1_3[output_df1_3.columns.difference(['CountryCode', 'Result'])].all().all():

    print("Check 1.3 CountryNames are same across all tables which are in demand: Pass")

else:

    print("Check 1.3 CountryNames are same across all tables which are in demand: Fail")

   
 
# Display the output DataFrame

#print(output_df1_3)
 
 
#Check1.4
 
# Define the list of DataFrames to check

dataframes_to_check = [df_demand, df_LabourData, df_RM_Supply, df]
 
# Function to check if 'MthNum' column contains values from 1 to 12

def check_mthnum_column(df):

    if 'MthNum' in df.columns:

        if all(1 <= df['MthNum']) and all(df['MthNum'] <= 12):

            return True

        else:

            return False

    else:

        return True
 
# Dictionary to store pass/fail results for each DataFrame

results = {}
 
# Check 'MthNum' column for each DataFrame

for df_name, df_to_check in zip(['df_demand', 'df_LabourData', 'df_RM_Supply', 'df'], dataframes_to_check):

    results[df_name] = check_mthnum_column(df_to_check)
 
# Generate the output DataFrame

output_df1_4 = pd.DataFrame(list(results.items()), columns=['DataFrame', 'Result'])
 
# Print Pass or Fail based on the check

if all(results.values()):

    print("Check 1.4 MthNum (Month numbers are from 1 to 12): Pass")

else:

    print("Check 1.4 MthNum (Month numbers are from 1 to 12): Fail")

    
# Display the output DataFrame

#print(output_df1_4)


# Check 2 (Forcasted Demand availability)
# Check if all values in 'DemandCases' column are greater than or equal to 0
df_QC_2 = df_demand.copy()
df_QC_2['Result'] = df_QC_2['DemandCases'] >= 0

all_demand_cases_positive = (df_QC_2['DemandCases'] >= 0).all()
print(f"Check 2: Forecasted Demand availability (All FGSKUs are having demand > 0): {'Pass' if all_demand_cases_positive else 'Fail'}")

# Check -3
df_Production=df_Production[df_Production['retain_flag']==1]
df_Production[['FactCode','FGSKUCode']].drop_duplicates()
df_BOM1 = df_BOM[df_BOM['BOMVersion'] == 'BOM-1']
df_BOM2 = df_BOM[df_BOM['BOMVersion'] == 'BOM-2']
# Extract unique combinations of 'FactCode' and 'FGSKUCode' from df_BOM1
bom1_unique_combinations = set(map(tuple, df_BOM[['FactCode', 'FGSKUCode']].drop_duplicates().values))

# Extract unique combinations of 'FactCode' and 'FGSKUCode' from df_Production
production_unique_combinations = set(map(tuple, df_Production[['FactCode', 'FGSKUCode']].drop_duplicates().values))

# Create a DataFrame to display the results
df_QC3 = pd.DataFrame(list(production_unique_combinations), columns=['FactCode', 'FGSKUCode'])
df_QC3['Result'] = [tuple(comb) in bom1_unique_combinations for comb in df_QC3[['FactCode', 'FGSKUCode']].values]

df_QC3.reset_index(drop=True, inplace=True)
if df_QC3['Result'].all():
    print("Check 3: All unique combinations from Production are present in BOM: Pass")
else:
    print("Check 3: All unique combinations from Production are not present in BOM1: Fail")

# Get unique combinations of 'FactCode' and 'FGSKUCode' from both DataFrames
unique_combinations_df1 = set(map(tuple, df_BOM1[['FactCode', 'FGSKUCode']].drop_duplicates().values))
unique_combinations_df2 = set(map(tuple, df_BOM2[['FactCode', 'FGSKUCode']].drop_duplicates().values))

# Create a DataFrame to display the results
df_QC4 = pd.DataFrame(list(unique_combinations_df2), columns=['FactCode', 'FGSKUCode'])
df_QC4['Result'] = [tuple(comb) in unique_combinations_df1 for comb in df_QC4[['FactCode', 'FGSKUCode']].values]

if df_QC4['Result'].all():
    print("Check-4: The combinations from BOM2 present in BOM1: Pass")
else:
    print("Check-4: The combinations from BOM2 present in BOM1: Fail")
    # Print combinations from BOM2 that are not present in BOM1
# Check if all FGSKUCode values from df_demand are present in df_Production
#all_fgsku_from_demand_in_production = df_demand['FGSKUCode'].isin(df_Production['FGSKUCode'])

# Create a DataFrame to display the results
df_QC5_1 = pd.DataFrame(df_demand['FGSKUCode'].unique(), columns=['FGSKUCode'])
df_QC5_1['Result'] = df_QC5_1['FGSKUCode'].isin(df_Production['FGSKUCode'])
print('Check 5.1: All FGSKU from demand are having mapping capacity (Demand consistency with Production):', "Pass" if df_QC5_1['Result'].all() else "Fail")

# Get unique FactCode and FGSKUCode combinations from BOM-1
unique_bom1_combinations = set(zip(df_BOM[df_BOM['BOMVersion'] == 'BOM-1']['FactCode'], df_BOM[df_BOM['BOMVersion'] == 'BOM-1']['FGSKUCode']))

# Get unique FactCode and FGSKUCode combinations from df_Production
unique_production_combinations = set(zip(df_Production['FactCode'], df_Production['FGSKUCode']))

# Create a DataFrame to display the results for Check 5.2
df_QC5_2 = pd.DataFrame(list(unique_bom1_combinations), columns=['FactCode', 'FGSKUCode'])
df_QC5_2['Result'] = [tuple(comb) in unique_production_combinations for comb in df_QC5_2[['FactCode', 'FGSKUCode']].values]

if df_QC5_2['Result'].all():
    print('Check 5.2: All Combinations of Factcode and FGSKU from BOM-1 have capacity mapping (BOM-1 consistency with Production): Pass')
else:
    print("Check 5.2: Some combinations of Factcode and FGSKU from BOM-1 do not have capacity mapping (BOM-1 consistency with Production): Fail")

# Get unique FactCode and FGSKUCode combinations from BOM-2
unique_bom2_combinations = set(zip(df_BOM[df_BOM['BOMVersion'] == 'BOM-2']['FactCode'], df_BOM[df_BOM['BOMVersion'] == 'BOM-2']['FGSKUCode']))

# Get unique FactCode and FGSKUCode combinations from df_Production
unique_production_combinations = set(zip(df_Production['FactCode'], df_Production['FGSKUCode']))

# Create a DataFrame to display the results for Check 5.3
df_QC5_3 = pd.DataFrame(list(unique_bom2_combinations), columns=['FactCode', 'FGSKUCode'])
df_QC5_3['Result'] = [tuple(comb) in unique_production_combinations for comb in df_QC5_3[['FactCode', 'FGSKUCode']].values]

if df_QC5_3['Result'].all():
    print('Check 5.3: All Combinations of Factcode and FGSKU from BOM-2 have capacity mapping (BOM-2 consistency with Production): Pass')
else:
    print("Check 5.3: Some combinations of Factcode and FGSKU from BOM-2 do not have capacity mapping (BOM-2 consistency with Production): Fail")

# Get unique 'FactCode' and 'LineCode' combinations from df_Production
production_combinations = set(zip(df_Production['FactCode'], df_Production['LineCode']))

# Get unique 'FactCode' and 'LineCode' combinations from df_lines
lines_combinations = set(zip(df_lines['FactCode'], df_lines['LineCode']))

# Get unique 'FactCode' and 'LineCode' combinations from df_Factory_Operations
factory_operations_combinations = set(zip(df_Factory_Operations['FactCode'], df_Factory_Operations['LineCode']))

# Create a DataFrame to display the results for Check 6
df_QC6 = pd.DataFrame(list(production_combinations), columns=['FactCode', 'LineCode'])
df_QC6['Result_lines'] = [tuple(comb) in lines_combinations for comb in df_QC6[['FactCode', 'LineCode']].values]
df_QC6['Result_factory_operations'] = [tuple(comb) in factory_operations_combinations for comb in df_QC6[['FactCode', 'LineCode']].values]
df_QC6['Result']= df_QC6['Result_lines'].all() and df_QC6['Result_factory_operations'].all()

if df_QC6['Result'].all():
    print("Check 6: Factory Operations data completeness (All combinations of FactCode and LineCode from Production are available in Factory Operations and Lines Master): Pass")
else:
    print("Check 6: Factory Operations data completeness (All combinations of FactCode and LineCode from Production are available in Factory Operations and Lines Master): Fail")
    # Print combinations from df_Production that are not present in either df_lines or df_Factory_Operations
    
# Check 7
# Check- Primary Freight Rates Availability

# Get unique combinations of 'FactCode' and 'FGSKUCode' from df_Production
unique_combinations_production = set(zip(df_Production['FactCode'], df_Production['FGSKUCode']))

# Get unique combinations of 'FACTCode' and 'FGSKUCode' from df_Primary_Freight
unique_combinations_primary_freight = set(zip(df_Primary_Freight['FACTCode'], df_Primary_Freight['FGSKUCode']))

# Create a DataFrame to store the results for Check 7.1
df_QC7_1 = pd.DataFrame(unique_combinations_production, columns=['Production_FactCode', 'Production_FGSKUCode'])
df_QC7_1['Result'] = [comb in unique_combinations_primary_freight for comb in unique_combinations_production]

if df_QC7_1['Result'].all():
    print("Check 7.1: All combinations from df_Production are present in df_Primary_Freight : Pass")
else:
    print("Check 7.1: All combinations from df_Production are present in df_Primary_Freight : Fail")

# Create a DataFrame to store the results for Check 7.2
df_QC7_2 = pd.DataFrame(unique_combinations_primary_freight, columns=['Pri_frt_FactCode', 'Pri_frt_FGSKUCode'])
df_QC7_2['Result'] = [comb in unique_combinations_production for comb in unique_combinations_primary_freight]

if df_QC7_2['Result'].all():
    print("Check 7.2: All combinations from Primary Freight are present in Production : Pass")
else:
    print("Check 7.2: All combinations from Primary Freight are present in Production : Fail")

print("Check 7.3: FrtCostPerUnit for NFC and NFCM should be 0:", "Pass" if all(df_Primary_Freight[(df_Primary_Freight['FACTCode'] == "NFC") & (df_Primary_Freight['WHCode'] == 'NFCM')]['FrtCostPerUnit'] == 0) else "Fail")

print("Check 7.4: FrtCostPerUnit for KFC and KFCM should be 0:", "Pass" if all(df_Primary_Freight[(df_Primary_Freight['FACTCode'] == "KFC") & (df_Primary_Freight['WHCode'] == 'KFCM')]['FrtCostPerUnit'] == 0) else "Fail")

print("Check 7.5: FrtCostPerUnit for GFC and GFCM should be 0:", "Pass" if all(df_Primary_Freight[(df_Primary_Freight['FACTCode'] == "GFC") & (df_Primary_Freight['WHCode'] == 'GFCM')]['FrtCostPerUnit'] == 0) else "Fail")


# Check 8
# Check - Secondary Freight Rates Availability

unique_fgsku_secondary_freight = set(df_Secondary_Freight['FGSKUCode'].unique())
unique_fgsku_demand = set(df_demand['FGSKUCode'].unique())

all_fgsku_secondary_in_demand = unique_fgsku_secondary_freight.issubset(unique_fgsku_demand)

countries_with_demand = df_demand['CountryCode'].unique()

countries_with_freight = df_Secondary_Freight.loc[df_Secondary_Freight['FGSKUCode'].isin(unique_fgsku_demand), 'CountryCode'].unique()

all_countries_with_demand_have_freight = set(countries_with_demand).issubset(set(countries_with_freight))

condition1 = (df_Secondary_Freight['CountryCode'] == 'Kuwait') & (df_Secondary_Freight['WHCode'] == 'KFCM')
condition2 = (df_Secondary_Freight['CountryCode'] == 'Saudi Arabia') & (df_Secondary_Freight['WHCode'] == 'NFCM')
condition3 = (df_Secondary_Freight['CountryCode'] == 'United Arab Emirates') & (df_Secondary_Freight['WHCode'] == 'GFCM')

result1 = (df_Secondary_Freight.loc[condition1, 'TransportCostPerCase'] == 0).all()
result2 = (df_Secondary_Freight.loc[condition2, 'TransportCostPerCase'] == 0).all()
result3 = (df_Secondary_Freight.loc[condition3, 'TransportCostPerCase'] == 0).all()

check_8_pass = all([all_fgsku_secondary_in_demand, all_countries_with_demand_have_freight, result1, result2, result3])
df_QC8_1 = pd.DataFrame(unique_fgsku_secondary_freight, columns=['Sec_frt_FGSKUCode'])
df_QC8_1['Result'] = [comb in unique_fgsku_demand for comb in unique_fgsku_secondary_freight]

df_QC8_2 = pd.DataFrame(countries_with_demand, columns=['Demand_Country'])
df_QC8_2['Result'] = [comb in countries_with_freight for comb in countries_with_demand]

print('Check 8: Secondary Freight Rates Checks :', "Pass" if check_8_pass else "Fail")

if not check_8_pass:
    # Print combinations from df_Secondary_Freight that are not present in df_demand
    missing_fgsku = unique_fgsku_secondary_freight - unique_fgsku_demand
    print("FGSKUCode from df_Secondary_Freight not present in df_demand:", missing_fgsku)

print("Check 8.1: All unique FGSKUCode values from Secondary Freight are present in Demand:", "Pass" if all_fgsku_secondary_in_demand else "Fail")

print("Check 8.2: All countries with demand have freight values from at least one warehouse:", "Pass" if all_countries_with_demand_have_freight else "Fail")

print("Check 8.3: TransportCostPerCase from KFCM to Kuwait should be 0:", "Pass" if result1 else "Fail")
print("Check 8.4: TransportCostPerCase from NFCM to Saudi Arabia should be 0:", "Pass" if result2 else "Fail")
print("Check 8.5: TransportCostPerCase from GFCM to United Arab Emirates should be 0:", "Pass" if result3 else "Fail")


#Check 9
 
# Assume dataframes is a list containing your DataFrames df_fgsku, df_Secondary_Freight, etc.

dataframes = [df_fgsku, df_Secondary_Freight, df_Primary_Freight, df_Production, df_BOM, df_FG_M1OS]
 
# Get unique FGSKUCode values from df_demand

unique_fgsku_codes_base = set(df_demand['FGSKUCode'].unique())
 
# Create a dictionary to store the results

qc9_data = {'FGSKUCode': list(unique_fgsku_codes_base)}
 
# Iterate through the list of DataFrames

for dataframe_name, dataframe in zip(['df_fgsku', 'df_Secondary_Freight', 'df_Primary_Freight', 'df_Production', 'df_BOM', 'df_FG_M1OS'], dataframes):

    # Check if 'FGSKUCode' column exists in the current DataFrame

    if 'FGSKUCode' in dataframe.columns:

        # Get unique FGSKUCode values from the current DataFrame

        unique_fgsku_codes_current = set(dataframe['FGSKUCode'].unique())

        # Compare the unique FGSKUCode values with those of df_demand

        qc9_data[dataframe_name] = [fgsku in unique_fgsku_codes_current for fgsku in qc9_data['FGSKUCode']]

    else:

        # If 'FGSKUCode' column does not exist in the current DataFrame, mark all entries as False

        qc9_data[dataframe_name] = [False] * len(qc9_data['FGSKUCode'])
 
# Create df_QC9 DataFrame from the dictionary

df_QC9 = pd.DataFrame(qc9_data)
 
# Check if all columns except 'FGSKUCode' are True for each row

df_QC9['Result'] = df_QC9.iloc[:, 1:].all(axis=1)
 
if df_QC9['Result'].all():

     print("Check 9: FGSKUCode are the same across all DataFrames: Pass")

else:

    print("Check 9: FGSKUCode are the same across all DataFrames: Fail")
 
# Print the results

#df_QC9

#check 10.1
# Get unique RMSKUCode values from each DataFrame
unique_rmsku_codes_df_rmsku = set(df_rmsku['RMSKUCode'].unique())
unique_rmsku_codes_df_BOM = set(df_BOM['RMSKUCode'].unique())
unique_rmsku_codes_df_RM_M1OS = set(df_RM_M1OS['RMSKUCode'].unique())

# Create separate DataFrames for each table
df_QC10_1 = pd.DataFrame(unique_rmsku_codes_df_rmsku, columns=['RMSKU'])
df_QC10_1['BOM'] = [rmsku in unique_rmsku_codes_df_BOM for rmsku in unique_rmsku_codes_df_rmsku]
df_QC10_1['RM_M1OS'] = [rmsku in unique_rmsku_codes_df_RM_M1OS for rmsku in unique_rmsku_codes_df_rmsku]
df_QC10_1['Result'] = df_QC10_1[['BOM', 'RM_M1OS']].all(axis=1)

# Reorder columns
df_QC10_1 = df_QC10_1[['RMSKU', 'BOM', 'RM_M1OS', 'Result']]
if df_QC10_1['Result'].all():
    print("Check 10.1: RMSKUCode are the same across all DataFrames : Pass")
else:
        print("Check 10.1: RMSKUCode are the same across all DataFrames : Fail")
# Print the result
#df_QC10_1

# Check 10.2

# Step 1: Select relevant columns and drop duplicates
df_BOM_unique = df_BOM[['FactCode','RMSKUCode']].drop_duplicates()
df_RM_Supply_unique = df_RM_Supply[['FactCode','RMSKUCode','RMCostPerUnit']].drop_duplicates()

# Step 2: Perform an outer join on 'FactCode'
df_QC10_2 = pd.merge(df_BOM_unique, df_RM_Supply_unique, on=['FactCode','RMSKUCode'], how='outer')

# Step 3: Filter out rows where 'RMCostPerUnit' is null
df_QC10_2['Result'] = df_QC10_2['RMCostPerUnit'].notnull()

if df_QC10_2['Result'].all():
    print("Check 10.2: All RMs having landed cost: Pass")
else:
    print("Check 10.2: All RMs having landed cost :Fail")

    
#Check 11
df_QC11_1 = df_rm_transfer.copy()
df_QC11_1['Result'] = df_QC11_1.apply(lambda x: 'True' if x['OrigFact'] != x['DestFact'] else 'False', axis=1)
df_QC11_1= df_QC11_1[['OrigFact', 'DestFact', 'Result']]
if df_QC11_1['Result'].all():
    print("Check 11.1: RM transfer Consistency : Pass")
else:
    print('Check 11.1: RM transfer Consistency : Fail')


df_QC11_2= df_rm_transfer.copy()
df_QC11_2['Result']= (df_rm_transfer['FrtRate'] != 0)
if df_QC11_2['Result'].all():
    print("Check 11.2: Trimming Freight is non zero : Pass")
else:
    print("Check 11.2: Trimming Freight cost is non zero : Fail")

# Check 12
# Chek 12.1 Chicken Brest in BOM-2
df_QC_12_1=df_BOM[df_BOM['BOMVersion']=='BOM-2']
df_QC_12_1['Result'] = df_QC_12_1['RMSKUCode'].astype(str).apply(lambda x: False if '1001101108' in x else True)

if df_QC_12_1['Result'].all():
    print("Check 12.1: Chicken Breast in BOM-2 : Pass")
else:
    print("Check 12.1: Chicken Breast in BOM-2 : Fail")

# Check 12.2
df_QC_12_2=df_BOM[df_BOM['RMSKUCode'].astype(str).str.contains('1001101108')]
df_QC_12_2.groupby(['BOMVersion','FactCode','FGSKUCode','RMSKUCode'])['KGPerUnit'].sum().reset_index()

df_QC_12_2=df_BOM.groupby(['BOMVersion','FactCode','FGSKUCode','RMSKUCode'])['KGPerUnit'].sum().reset_index()
df_QC_12_2=df_QC_12_2[df_QC_12_2['RMSKUCode'].astype(str).str.contains('1001101108')]

# FG More than 1 Factory
FGs=df_QC_12_2.groupby(['BOMVersion','FGSKUCode'])['FactCode'].count().reset_index()
FGs=FGs[FGs['FactCode']!=1]             ###### Fact Count 
FGs=FGs.rename(columns={
    'FactCode':'FactCode_Count'
})

# FG with multiple Factory for 1001101108
df_QC_12_2=pd.merge(df_QC_12_2,FGs,on=['BOMVersion','FGSKUCode'])
df_QC_12_2['KGPerUnitMean'] = df_QC_12_2.groupby(['BOMVersion', 'FGSKUCode'])['KGPerUnit'].transform('mean')

# df_QC_12_2['% Diff']
df_QC_12_2['% Diff']=((df_QC_12_2['KGPerUnit']-df_QC_12_2['KGPerUnitMean'])/df_QC_12_2['KGPerUnitMean'])*100
df_QC_12_2['Result'] = df_QC_12_2['% Diff'].apply(lambda x: True if -5 <= x <= 5 else False)

if df_QC_12_2['Result'].all():
    print("Check 12.2: Difference in chicken breast consumption between factories : Pass")
else:
    print("Check 12.2: Difference in chicken breast consumption between factories : Fail")
```
